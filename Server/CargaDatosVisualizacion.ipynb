###  Importar Librerias

```python
import numpy as np
import pandas as pd
import chardet
import json
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

with open('DomesticViolenceSpain.csv', 'rb') as f:
    result = chardet.detect(f.read())
    
data = pd.read_csv('DomesticViolenceSpain.csv', encoding=result['encoding'])
```

```python
print(data.values.shape)
data
```

```python
CantidadVictimas = data.groupby([ "Year"] ).size().to_frame(name = 'Victims').reset_index()
CantidadVictimas.sort_values(['Year'], ascending=True)
```

```python
Victimas = pd.DataFrame(CantidadVictimas,columns = ['Year','Victims'])
Victimas.shape
```

```python
Victimas
```

```python

plt.figure(figsize=(10,6))
g = sn.factorplot("Year", "Victims", data=Victimas, kind="bar", size=10, palette="muted", legend=False)
plt.show()
```

```python
Victimas.to_csv('Victims.csv',header= False)
```
